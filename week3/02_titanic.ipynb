タイタニック号沈没事故における性別ごとの乗客生存率を求めてみましょう。  
データセットの読み込み方法は以下で解説します。  
性別に関する情報は`'Sex'`列(男性:`'male'`、女性:`'female'` のみ)、生存情報は`'Survived'`列(生存者:`1`、非生存者:`0`)に記載されています。  
生存率$R$は以下のように計算することとする

$
R = N_{suv}/N_{all}
$

```python
# from google.colab import drive
# drive.mount('/content/drive')

import pandas as pd

# Titanicデータセットを読み込む
# titanic_df = pd.read_csv('/content/drive/MyDrive/titanic.csv') # 先ほど取得したpathに書き換えて下さい
titanic_df = pd.read_csv('/tmp/work/git/ai-cource-practice/week3/titanic.csv') # 先ほど取得したpathに書き換えて下さい

titanic_df
```

```python
### 以下を変更して下さい ###

# 男性の生存率を求める
N_m_all = len(titanic_df[titanic_df['Sex']=='male'])
N_m_suv = len(titanic_df[(titanic_df['Sex']=='male') & (titanic_df['Survived'] == 1)])
male_survival_rate = N_m_suv/N_m_all

# 女性の生存率を求める
N_f_all = len(titanic_df[titanic_df['Sex']=='female'])
N_f_suv = len(titanic_df[(titanic_df['Sex']=='female') & (titanic_df['Survived'] == 1)])
female_survival_rate = N_f_suv/N_f_all


print(f"男性の生存率: {male_survival_rate:.3f}")
print(f"女性の生存率: {female_survival_rate:.3f}")
```

```python
N_m_all = len(titanic_df[titanic_df['Sex']=='male'])
N_f_all = len(titanic_df[titanic_df['Sex']=='female'])

N_m_suv = len(titanic_df[(titanic_df['Sex']=='male') & (titanic_df['Survived'] == 1)])# 条件式は丸カッコで囲もう！
N_f_suv = len(titanic_df[(titanic_df['Sex']=='female') & (titanic_df['Survived'] == 1)])# 条件式は丸カッコで囲もう！

print(N_m_all)
print(N_f_all)
print(N_m_suv)
print(N_f_suv)

R_m = N_m_suv/N_m_all
R_f = N_f_suv/N_f_all

print(R_m)
print(R_f)

```
